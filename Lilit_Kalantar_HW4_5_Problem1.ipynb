```
# Importing libraries. 
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import re
import requests
import scrapy
from scrapy.http import TextResponse
from itertools import combinations
```

```
# Finding the chromedriver. 
PATH = r"/Users/apple/Desktop/chromedriver"
options = Options()
options.headless = True
URL = "http://rate.am/en/armenian-dram-exchange-rates/central-bank-armenia"
browser = webdriver.Chrome(PATH)
browser.get(URL)
time.sleep(1)
```

```
# I created serparate empty lists for each month and days of 2018, 2019, and 2020. 
# To get the data from the table, I scraped days and months using xpath.
# After scraping and filtering the data, it is added to those lists and then combines in a DataFrame.
# I chose USD, so I got 3 DataFrames in the end: USD_2018, USD_2019 and USD_2020.  
Months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
Years = ["2018", "2019", "2020"]
for i in Years:
    year = browser.find_element_by_xpath(f'//select[@name="ctl00$Content$dlYear"]/option[@value={i}]')
    year.click()
    time.sleep(1)
    page = browser.page_source
    response = TextResponse(body = page, url = URL, encoding = "utf-8")
    if response.xpath(f'//select[@name="ctl00$Content$dlYear"]/option[@selected="selected"][@value="{i}"]'):
        Days_2018 = []
        January_2018 = []
        February_2018 = []
        March_2018  = []
        April_2018 = []
        May_2018 = []
        June_2018 = []
        July_2018 = []
        August_2018 = []
        September_2018 = []
        October_2018 = []
        November_2018 = []
        December_2018 = []
        Days_2019 = []
        January_2019 = []
        February_2019 = []
        March_2019 = []
        April_2019 = []
        May_2019 = []
        June_2019 = []
        July_2019 = []
        August_2019 = []
        September_2019 = []
        October_2019 = []
        November_2019 = []
        December_2019 = []
        Days_2020 = []
        January_2020 = []
        February_2020 = []
        March_2020 = []
        April_2020 = []
        May_2020 = []
        June_2020 = []
        July_2020 = []
        August_2020 = []
        September_2020 = []
        October_2020 = []
        November_2020 = []
        December_2020 = []
        
        for j in range (2,33):

            days = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[1]/text()").extract()
            days = days[0].strip()
            january_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[2]/text()").extract()
            january = ' '.join([str(elem) for elem in january_draft])
            january = january.strip()
            february_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[3]/text()").extract()
            february = ' '.join([str(elem) for elem in february_draft)
            february = february.strip()
            march_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[4]/text()").extract()
            march = ' '.join([str(elem) for elem in march_draft])
            march = march.strip()
            april_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[5]/text()").extract()
            april = ' '.join([str(elem) for elem in april_draft])
            april = april.strip()
            may_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[6]/text()").extract()
            may = ' '.join([str(elem) for elem in may_draft])
            may = may.strip()
            june_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[7]/text()").extract()
            june = ' '.join([str(elem) for elem in june_draft])
            june = june.strip()
            july_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[8]/text()").extract()
            july = ' '.join([str(elem) for elem in july_draft])
            july = july.strip()
            august_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[9]/text()").extract()
            august = ' '.join([str(elem) for elem in august_draft])
            august = august.strip()
            september_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[10]/text()").extract()
            september = ' '.join([str(elem) for elem in september_draft])
            september = september.strip()
            october_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[11]/text()").extract()
            october = ' '.join([str(elem) for elem in october_draft])
            october = october.strip()
            november_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[12]/text()").extract()
            november = ' '.join([str(elem) for elem in november_draft])
            november = november.strip()
            december_draft = response.xpath(f"//table[@class='cb']/tbody/tr[{j}]/td[13]/text()").extract()
            december = ' '.join([str(elem) for elem in december_draft])
            december = december.strip()
            
            if i == "2018":
                Days_2018.append(days)
                January_2018.append(january)
                February_2018.append(february)
                March_2018.append(march)
                April_2018.append(april)
                May_2018.append(may)
                June_2018.append(june)
                July_2018.append(july)
                August_2018.append(august)
                September_2018.append(september)
                October_2018.append(october)
                November_2018.append(november)
                December_2018.append(december)
                USD_2018 = pd.DataFrame({"Days": Days_2018, "January": January_2018, "February": February_2018, "March": March_2018, "April": April_2018, "May": May_2018, "June": June_2018, "July": July_2018, "August": August_2018, "September": September_2018, "October": October_2018, "November": November_2018, "December": December_2018})
                
            if i == "2019":
                Days_2019.append(days)
                January_2019.append(january)
                February_2019.append(february)
                March_2019.append(march)
                April_2019.append(april)
                May_2019.append(may)
                June_2019.append(june)
                July_2019.append(july)
                August_2019.append(august)
                September_2019.append(september)
                October_2019.append(october)
                November_2019.append(november)
                December_2019.append(december)
                USD_2019 = pd.DataFrame({"Days": Days_2019, "January": January_2019, "February": February_2019, "March": March_2019, "April": April_2019, "May": May_2019, "June": June_2019, "July": July_2019, "August": August_2019, "September": September_2019, "October": October_2019, "November": November_2019, "December": December_2019})
                
            if i == "2020":
                Days_2020.append(days)
                January_2020.append(january)
                February_2020.append(february)
                March_2020.append(march)
                April_2020.append(april)
                May_2020.append(may)
                June_2020.append(june)
                July_2020.append(july)
                August_2020.append(august)
                September_2020.append(september)
                October_2020.append(october)
                November_2020.append(november)
                December_2020.append(december)
                USD_2020 = pd.DataFrame({"Days": Days_2020, "January": January_2020, "February": February_2020, "March": March_2020, "April": April_2020, "May": May_2020, "June": June_2020, "July": July_2020, "August": August_2020, "September": September_2020, "October": October_2020, "November": November_2020, "December": December_2020})

browser.close()
```

```
# Cleaning and converting my data for answering the questions. 
USD_2018_final = USD_2018.replace("X", "NaN").replace("", "NaN")
USD_2019_final = USD_2019.replace("X", "NaN").replace("", "NaN")
del USD_2018_final['Days']
for i in USD_2018_final:
    USD_2018_final[f"{i}"] = pd.to_numeric(USD_2018_final[f"{i}"], downcast = "float", errors='coerce')
del USD_2019_final['Days']
for i in USD_2019_final:
    USD_2019_final[f"{i}"] = pd.to_numeric(USD_2019_final[f"{i}"], downcast = "float", errors='coerce')
for i in Months:
    m_to_m_dif = USD_2019_final.sub(USD_2019_final, axis = i)
```

```
# 1: Calculate annual variation both for 2018 and 2019. 
# I used var() function to get the variations. 
# var_2018: almost 1.153459
# var_2019: almost 1.153459
var_2018 = str(USD_2018_final.var().mean())
var_2019 = str(USD_2018_final.var().mean())
```
